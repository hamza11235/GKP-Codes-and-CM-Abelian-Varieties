# Phase 4: matched generic non-CM controls

Each exact CM record is compared with three small generic-real deformations at fixed polarization matrix, fixed type $D$, and fixed metric normalization. The deformations use rational multiples of $\pi$, so they leave the rational CM/isogeny orbit. Generic members are non-CM almost surely; this notebook does **not** certify the endomorphism ring of each individual control.

```python
import sys
from pathlib import Path

here = Path.cwd().resolve()
candidates = (here, here.parent, here / 'passive-cliffords')
project = next(path for path in candidates if (path / 'src' / 'gkp_passive_cliffords').exists())
workspace = project.parent
sys.path.insert(0, str(project / 'src'))
sys.path.insert(0, str(workspace / 'src'))

from gkp_passive_cliffords import phase4_comparison_table, phase4_control_table
```

## Matched comparison

The enhancement factor is the CM logical-image order divided by the largest image order among its three matched controls.

```python
comparisons = phase4_comparison_table()
columns = [
    'candidate_id', 'polarization_type',
    'cm_ell_squared', 'control_ell_squared_min', 'control_ell_squared_max',
    'cm_automorphism_order', 'control_automorphism_orders',
    'cm_logical_image_order', 'control_logical_image_orders',
    'logical_image_enhancement',
]
print(' | '.join(columns))
print(' | '.join(['---'] * len(columns)))
for row in comparisons:
    print(' | '.join(str(row[column]) for column in columns))
```

## Individual control checks

A generic polarized metric retains the unavoidable origin-fixing automorphisms $\{\pm I\}$. At exponent two, $-I$ acts trivially on $K(L)$, so an order-two geometric group can have logical image order one.

```python
controls = phase4_control_table()
for row in controls:
    print(
        f"{row['candidate_id']:36s} #{row['control_index']}  "
        f"ell^2={row['ell_squared']:.9f}  "
        f"Aut={row['polarized_automorphism_order']}  "
        f"image={row['logical_image_order']}  "
        f"relative metric shift={row['relative_metric_displacement']:.3e}"
    )
```

## Result

All 18 controls have the minimal group $\{\pm I\}$. Relative to these fixed-$D$ local controls, the six CM records enhance the passive logical image by factors $12,12,3,6,12,$ and $48$, respectively. This is evidence for CM-enhanced passive Clifford symmetry in this sample, not a theorem about all CM points or all non-CM points.

```python
assert len(controls) == 18
assert all(row['polarized_automorphism_order'] == 2 for row in controls)
expected = [12, 12, 3, 6, 12, 48]
assert [row['logical_image_enhancement'] for row in comparisons] == expected
print('Phase 4 notebook checks passed.')
```
